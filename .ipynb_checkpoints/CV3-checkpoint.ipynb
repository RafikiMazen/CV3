```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from matplotlib.pyplot import imread
```

```python
def imOpen(image_input):
    original_image= Image.open(image_input)
    grey_image=original_image.convert("L")
    plt.imshow(original_image,cmap='gray')
    return grey_image
```

```python
def imgToArray(path):
    return imread(path)
```

```python
imOpen("BarCode2.jpg")
imgA = imgToArray("BarCode2.jpg")
```

```python
def applyWeightedToPixel(image,i,j):
    weightedMedian = np.array([[1, 3, 1],
                               [3, 5, 3],
                               [1, 3, 1]
                                     ])
    values=[]
    for ii in range(0,3):
        for jj in range(0,3):
            v = image[i+ii-1][j+jj-1]
            l = [v]*weightedMedian[ii][jj]
            values.extend(l)
    values=np.array(values)
    values=np.sort(values)
    return values[values.shape[0]//2]
    
    
```

```python
def applyNonWeightedToPixel(image,i,j):
    values=[]
    for ii in range(0,3):
        for jj in range(0,3):
            v = image[i+ii-1][j+jj-1]
            l = [v]
            values.extend(l)
    values=np.array(values)
    values=np.sort(values)
    return values[values.shape[0]//2]
    
```

```python
def applyMedianToImg(image,weighted):
    e=1
    rImg=np.zeros(image.shape)
    for i in range(e,image.shape[0]-e):
        for j in range(e,image.shape[1]-e):
            if(weighted):
                rImg[i][j]=applyWeightedToPixel(image,i,j)
            else:
                rImg[i][j]=applyNonWeightedToPixel(image,i,j)
                
    rImg = rImg.astype(int) 
    return rImg   
```

```python
print("WEIGHTED MEDIAN")
weightedImg = applyMedianToImg(imgA,True)
plt.imshow(weightedImg,cmap="gray")
img=Image.fromarray(weightedImg*255)
img.save("./BarCodeWeightedMedian.png")
```

```python
print("NON WEIGHTED MEDIAN")
nonWeightedImg = applyMedianToImg(imgA,False)
plt.imshow(nonWeightedImg,cmap="gray")
img=Image.fromarray(nonWeightedImg*255)
img.save("./BarCodeMedian.png")
```
